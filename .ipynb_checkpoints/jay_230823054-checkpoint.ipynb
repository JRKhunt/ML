```python
import pandas as pd
import plotly.express as px
```

```python
# Read the dataset and display the 10 records.

df=pd.read_excel("world_literacy_data_2021_2023.xlsx").head(10)
df
```

```python
# Print all the column names.

print(df.columns)
```

```python
# Print the size(rows, cols) of dataset.

df=pd.read_excel("world_literacy_data_2021_2023.xlsx")
print(df.size) 
```

```python
# Check for the NULL values in dataset and if found replace the NULL values with the average values.

print(df.isnull().sum())
df.fillna(df.mean(numeric_only=True),inplace=True)
df
```

```python
# Print the updated dataset.

print(df.head(10))
```

```python
# Display the top 10 countries having highest GDP year 2023.

Top_gdp_2023 = df[df['Year'] == 2023].sort_values(by='GDP',ascending=False).head(10)
print(Top_gdp_2023[['Country','GDP','Year']])
```

```python
# Display top 10 countries having highest Literacy rate year 2023.

Top_literacy_2023 = df[df['Year'] == 2023].sort_values(by='Literacy Rate', ascending=False).head(10)
print(Top_literacy_2023[['Country', 'Literacy Rate' , 'Year']])
```

```python
# Plot the year wise literacy rate of each country.

import pandas as pd
import plotly.express as px
df =  pd.read_excel("world_literacy_data_2021_2023.xlsx")
fig = px.line(df, x='Year', y='Literacy Rate', color='Country')
fig.update_layout(
    title='Year-wise Literacy Rate of Each Country',
    xaxis_title='Year',
    yaxis_title='Literacy Rate',
    legend_title='Country'
)
fig.show()
```

```python
# Plot the year wise GDP of each country.

import pandas as pd
import plotly.express as px
df =  pd.read_excel("world_literacy_data_2021_2023.xlsx")
fig_gdp = px.line(df, x='Year', y='GDP', color='Country', markers=True)
fig.update_layout(
     title='Year-wise GDP of Each Country',
    xaxis_title='Year',
    yaxis_title='GDP',
    legend_title='Country',
    legend=dict(x=1, y=1),  # Position legend outside the plot area
    margin=dict(l=0, r=200, t=40, b=40)  # Adjust margins to fit the legend
)
fig.show()
```

```python
# Plot the GDP and Literacy rate of the each country (Year : 2023).

import plotly.express as px
df_2023 = df[df['Year'] == 2023]
df = pd.read_excel('world_literacy_data_2021_2023.xlsx') 
fig = px.scatter(df_2023, x="GDP", y="Literacy Rate", color='Country', 
                                 title='GDP vs Literacy Rate of Each Country (Year: 2023)')
fig.update_layout(
    xaxis_title='GDP',
    yaxis_title='Literacy Rate',
    legend_title='Country'
)
fig.show()
```
