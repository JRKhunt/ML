```python
# blur

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("modi.webp")
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

blur_img = cv2.blur(new_img,(3,3))
plt.imshow(blur_img)
```

```python
# box blur

img = cv2.imread("modi.webp")
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blurred_img1 = cv2.blur(new_img,(8,8))

plt.subplot(1,2,1),plt.imshow(new_img),plt.title("original image")
plt.axis('off')
plt.subplot(1,2,2),plt.imshow(blurred_img1),plt.title("box image")
plt.axis('off')
```

```python
# sharpning

import numpy as np

img = cv2.imread("modi.webp")
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
k1 = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
k2 = np.array([[1,1,1],[1,-7,1],[1,1,1]])

new_img1 = cv2.filter2D(img,-1,k1)
new_img2 = cv2.filter2D(img,-1,k2)
plt.subplot(1,3,1),plt.imshow(new_img)
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(new_img1)
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(new_img2)
plt.axis('off')
```

```python
# edge detection

sobelx = cv2.Sobel(new_img,cv2.CV_64F,1,0,5)
sobely = cv2.Sobel(new_img,cv2.CV_64F,0,1,5)

plt.subplot(1,3,1),plt.imshow(new_img,cmap="gray")
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(sobelx,cmap="gray")
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(sobely,cmap="gray")
plt.axis('off')
```

```python
# face detection

import cv2
img = cv2.imread("3.jpg",cv2.IMREAD_COLOR)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(grey_img,1.1,4)
print(faces)
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow("face detector",img)
cv2.waitKey(50000)
cv2.destroyAllWindows()
```
