```python
# print all the records of data frame in alphabetical orders.

file_path = 'MOCK_DATA.csv'
df = pd.read_csv(file_path)
sorted_df = df.sort_values(by='Name') 
sorted_df
```

```python
# print all the records of dataframe in a way that the students having highest marks display the first.

df['Total'] = df['Marks1'] + df['Marks2'] + df['Marks3'] + df['Marks4'] 
df
df_sorted = df.sort_values(by='Total', ascending=False)
print(df_sorted.to_string(index=False))
```

```python
# add column "percentage" in above dataframe and assign the calculated percentage based on marks.

df['Percentage'] = df['Total']/4
df
```

```python
# print the details of students who lives in rajkot.

df_rajkot = df[df['City'] == 'Rajkot']
print(df_rajkot.to_string(index=False))
```

```python
# print the students who achieve the marks < 35.

df_low_marks = df[(df[['Marks1', 'Marks2', 'Marks3', 'Marks4']] < 35).any(axis=1)]
print(df_low_marks.to_string(index=False))
```

```python
# print the students whose name starts with 'M'.

df_names_starting_with_m = df[df['Name'].str.startswith('M')]
print(df_names_starting_with_m.to_string(index=False))
```

```python
# find the student having maximum percentage.

max_percentage_index = df['Percentage'].idxmax()
student_max_percentage = df.loc[max_percentage_index]
print(student_max_percentage.to_frame().T.to_string(index=False))
```

```python
# find the student having minimum percentage.

min_percentage_index = df['Percentage'].idxmin()
student_min_percentage = df.loc[min_percentage_index]
print(student_min_percentage.to_frame().T.to_string(index=False))
```

```python
# print the students who study in BCA and MCA.

streams_of_interest = ['Bca', 'Mca']
df_bca_mca = df[df['Stream'].isin(streams_of_interest)]
print(df_bca_mca.to_string(index=False))
```
