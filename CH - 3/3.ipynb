```python
# kmeans algorithm : unsupervised learning(clustering)

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# step : 1 load the data
plt.figure()
data = pd.read_csv("Book1.csv")
x = data.iloc[:,0]
y = data.iloc[:,1]
plt.scatter(x,y)

# step : 2 load the model
model=KMeans(init="k-means++",n_clusters=4)

# step : 3 train the model
model.fit(data)

# use the model to diaplay the clusters
result = model.predict(data)
print(result)

# find the centroids (center point)
centroids = model.cluster_centers_
print(centroids)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import KMeans
import pandas as pd

# Load the CSV file
data = pd.read_csv("Book1.csv")
x = data.iloc[:, 0]
y = data.iloc[:, 1]

# Plot the dataset to identify the clusters
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.title('Input data')
plt.xticks(())
plt.yticks(())

# Load the Model
kmeans = KMeans(init='k-means++', n_clusters=4)

# Train the Model
kmeans.fit(data)

# Use the Model to predict values
results = kmeans.predict(data)
print(results)

# Find the centroids
centroids = kmeans.cluster_centers_
print(centroids)

# Plot the clusters along with centroids
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='r')
plt.title('Centroids on boundaries obtained using KMeans')
plt.show()
```

```python
#  meanshift clustring

from sklearn.cluster import MeanShift, estimate_bandwidth
import pandas as pd

# Load the CSV file
data = pd.read_csv("2017.csv")

# preprocess
data = data.iloc[:,1:]

# calculate the bandwidth
bandwidth = estimate_bandwidth(data)

# load the model
meanshift = MeanShift(bandwidth = bandwidth)

# train the model
result = meanshift.fit(data)

# show the results
result.predict(data)
```

```python
# pip install scikit-image

from skimage import io 
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

image = io.imread("hqdefault.jpg")

# preprocessing : normalization
img_new = (image/255).reshape(-1,3)

# load the model 
img_kmeans = KMeans(n_clusters = 4)

# train the model 
img_kmean = img_kmeans.fit(img_new)

# rearrange the image using the results
img_rearranged = img_kmean.cluster_centers_[img_kmean.labels_]

img_reshaped = np.reshape(img_rearranged,(image.shape))
img.imsave("new.jpg",img_reshaped)
```
